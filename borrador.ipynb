```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
from sklearn.datasets import load_sample_image
from sklearn.datasets import load_digits

import warnings
warnings.filterwarnings("ignore")
```

```python
digitos = load_digits()
data = digitos.data

print(data.shape)
```

```python
data=255-data
print(data.shape)
```

```python
np.random.seed(1)
n = 10
kmean_ = KMeans(n_clusters=n, random_state=0, init='random', tol=0.00000000000000001, algorithm='elkan')
kmean_.fit(data)

yhat = kmean_.predict(data)
```

```python
for i in range(0,n):
    fila = np.where(yhat == i)[0]
    #print(fila)
    num = fila.shape[0]
    #print(num)
    r = int(np.floor(num/10))
    print("cluster: ", str(i))
    print(str(num) + " elementos")
    plt.figure(figsize=(10, 10))
    for j in range(0,num):
        plt.subplot(r+1, 10, j+1)
        plt.imshow(data[fila[j], :].reshape(8,8), cmap='gray')
        plt.axis('off')
    plt.show()
```
